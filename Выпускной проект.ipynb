# Задача

Построить модель, которая предскажет температуру стали во время последнего замера температуры, чтобы уменьшить потребление электроэнергии на этапе обработки стали


### Описание этапа обработки

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»).

### Первичная предобработка

Пропусков нет, типы в столбцах *Начало нагрева дугой* и	*Конец нагрева дугой* нужно привести к дата-время

```python
data_arc['Начало нагрева дугой'] = pd.to_datetime(data_arc['Начало нагрева дугой'],format='%Y-%m-%d %H:%M:%S')
data_arc['Конец нагрева дугой'] = pd.to_datetime(data_arc['Конец нагрева дугой'],format='%Y-%m-%d %H:%M:%S')
```

Дано время начала и конца нагрева (партии нагревают несколько раз),активная и реактивная мощности электрода. Активная мощность – это та самая мощность, потребляя которую, электронагреватель и вырабатывает тепловую энергию.  Добавим столбец *Длительность нагрева в секундах*

```python
data_arc['Длительность нагрева в секундах'] = (data_arc['Конец нагрева дугой'] - data_arc['Начало нагрева дугой'])/ np.timedelta64(1, 's')
data_arc
```

## 1. data_temp

Из этой таблицы мы возьмем: 
- Целевой признак - результат последнего измерения. 

(при этом проверим, что этот результат у нас есть (иначе будет просто невозможно обучать модели или проверять их предсказания), и в дальнейшем будем собирать данные только для key, у которых есть этот признак)

- Время первого и последнего замера - в качестве вспомогательных. Они помогут в отборе остальных признаков - чтобы собирать только те данные, которые происходили во время между этими замерами). Перед обучением модели мы их удалим.

- результат первого измерения температуры,
- длительность времени между первым и последним замерами,
- общее количество замеров. 

```python
###data_temp
# 'Время последнего замера', 'Последняя температура'
data_temp = data_temp.sort_values(by = 'Время замера', ascending = False)
last_temp = data_temp.drop_duplicates(subset=['key']).sort_values(by = 'key')
last_temp = last_temp.dropna().reset_index(drop = True)
last_temp.columns = ['key', 'Время последнего замера', 'Последняя температура']

# 'Время первого замера', 'Первая температура'
data_temp = data_temp.sort_values(by = 'Время замера')
first_temp = data_temp.drop_duplicates(subset=['key']).sort_values(by = 'key') 
first_temp.columns = ['key', 'Время первого замера', 'Первая температура']

#создаем основную таблицу
key_info = pd.DataFrame()
key_info = last_temp.merge(first_temp, how = 'left', on = 'key')

#'Между первым и последним замером'
temp_timedif = (pd.pivot_table(data_temp[['key', 'Время замера']], index='key',  aggfunc='max') - 
                pd.pivot_table(data_temp[['key', 'Время замера']], index='key',  aggfunc='min'))/np.timedelta64(1, 's')
temp_timedif.columns = ['Между первым и последним замером']
key_info = key_info.merge(temp_timedif, how = 'left', on = 'key')

#'Количество замеров'
total_temp_amount = pd.pivot_table(data_temp[['key', 'Температура']],index='key',  aggfunc='count')
total_temp_amount.columns = ['Количество замеров']
key_info = key_info.merge(total_temp_amount, how = 'left', on = 'key')

key_info.head(10)
```

В эту таблицу могли попасть некорректные данные - где было известно всего 1 значение температуры - мы не знаем, это первый замер или последний, таких записей всего 2:

# 3. Обучение различных моделей и выбор лучшей.

Сама по себе задача машинного обучения не настолько сложная, чтобы использовать сложные архитектуры типа нейронных сетей. Обучим 3 простые модели с различными гиперпараметрами - 

- DecisionTreeRegressor (проварьируем максимальную глубину), 
- RandomForestRegressor (проварьируем количество деревьев), 
- LinearRegression, 
- Для модели градиентного бустинга LightGBM проварьируем гипермараметры num_leaves, num_boost_round и early_stopping_rounds

на различном объеме признаков: 
- long будет включать все признаки, которые нам удалось собрать, 
- mid - long без отдельного выделения bulk i и wire i, 
- short - mid без параметров, связанных друг с другом, (например, из 'Среднее А:Р', 'Среднее А', 'Среднее Р' оставим только 'Среднее А:Р').

Сразу отделим тестовую выборку из исходного набора данных (20%) - на ней проверим победившую модель. Выбирать лучшую же будем по MAE на кросс-валидации, так как примеров у нас останется не так много - около 2000.

```python
train, test = train_test_split(data, test_size=0.2, random_state = 123)

# функция для разделения признаков (в указанном объеме - how) и целевого признака
def create_features_target(how, data):
    
    if (how == "long"):
        mydata = data
    else:
        mydata = data.drop(['Wire 1', 'Wire 2','Wire 3','Wire 4','Wire 5','Wire 6','Wire 7', 'Wire 8', 'Wire 9',
                           'Bulk 1', 'Bulk 2','Bulk 3','Bulk 4','Bulk 5','Bulk 6','Bulk 7','Bulk 8','Bulk 9','Bulk 10',
                           'Bulk 11','Bulk 12','Bulk 13','Bulk 14','Bulk 15'], axis = 1)
        if (how == "short"):
            mydata = mydata.drop(['between_last_n_first_temp', 'total_heat_time',
                                  'mean a', 'mean r', 'Bulk_amount', 'Wire_amount'], axis = 1) 
    features = mydata.drop('last_temp', axis = 1)
    target = mydata['last_temp']
    return features, target
```

```python
#здесь будем хранить результаты
compare = []
columns = ['features_vol', 'model', 'model_params', 'cross_valid_mae']

#начинаем сравнение
for features_vol in ["long", "mid", "short"]: 
        features, target = create_features_target(features_vol, train)
        
        #DecisionTreeRegressor
        for max_depth in range(1, 10, 4):
            model = DecisionTreeRegressor(max_depth = max_depth, random_state = 123, criterion = 'mae')
            mean_mae = (-1)*pd.Series(cross_val_score(model, features, target,scoring = 'neg_mean_absolute_error', cv=5)).mean()
            print(mean_mae)
            compare.append([features_vol, 'Tree', 'max_depth = {}'.format(max_depth), mean_mae])
        print()
        
        #Forest
        for n_estimators in range(10, 51, 20):
            model = RandomForestRegressor(n_estimators = n_estimators, random_state = 123, criterion = 'mae')
            mean_mae = (-1)*pd.Series(cross_val_score(model, features, target,scoring = 'neg_mean_absolute_error', cv=5)).mean()
            print(mean_mae)
            compare.append([features_vol, 'Forest', 'n_estimators = {}'.format(n_estimators), mean_mae])
        print()
            
        #LinearRegression
        model = LinearRegression(n_jobs = -1)
        mean_mae = (-1)*pd.Series(cross_val_score(model, features, target,scoring = 'neg_mean_absolute_error', cv=5)).mean()
        print(mean_mae)
        compare.append([features_vol, 'LinearRegression', '-', mean_mae])
        print()
```

```python
#Для модели градиентного бустинга LightGBM проварьируем гипермараметры num_leaves, num_boost_round и early_stopping_rounds

params = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': 'mean_absolute_error',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0
}

train, valid = train_test_split(train, test_size=0.2, random_state = 123)

for features_vol in ["long", "mid", "short"]:
    print(features_vol)
    features_train, target_train = create_features_target(features_vol, train)
    features_valid, target_valid = create_features_target(features_vol, valid)
    
    features_train = features_train.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
    target_train = target_train.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
    features_valid = features_valid.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
    target_valid = target_valid.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
    
    # create dataset for lightgbm
    lgb_train = lgb.Dataset(features_train, target_train, free_raw_data=False)
    lgb_valid = lgb.Dataset(features_valid, target_valid, reference=lgb_train, free_raw_data=False)

    for num_leaves0 in (20,40):
        for num_boost_round0 in (50, 100, 150):
            for early_stopping_rounds0 in (5, 10):
                params['num_leaves'] = num_leaves0

                gbm = lgb.train(params,
                    lgb_train,
                    num_boost_round=num_boost_round0,
                    valid_sets=lgb_valid,
                    early_stopping_rounds=early_stopping_rounds0)

                y_pred = gbm.predict(features_valid, num_iteration=gbm.best_iteration)


                mae = mean_absolute_error(target_valid, y_pred)
                print(mae)

                compare.append([features_vol, 'LIGHT', 'n_leaves = {}, n_boost = {}, rounds = {}'.format(
                    num_leaves0, num_boost_round0, early_stopping_rounds0), mae])
              
```

```python
#выводим результаты            
results = pd.DataFrame(data = compare, columns = columns)
results.sort_values('cross_valid_mae', ascending = True).head(50)
```
